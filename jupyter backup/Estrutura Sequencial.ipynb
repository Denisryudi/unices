#### 8. Vamos criar um conversor de temperatura. Faça um Programa que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius.
$C = \frac{5}{9}(F-32)$

```python
temp_fahrenheit = float(input("Enter a tempeture in Fahrenheit: "))
ceusius = 5/9 * (temp_fahrenheit - 32)

print(f"The temperature in Celsius is: {ceusius}")
```

#### 9. Faça um Programa que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.
$F = \frac{9}{5}C + 32$

```python
temp_celsius = float(input("Enter a tempeture in Celsius: "))
fahrenheit = ((9/5 * temp_celsius) + 32)

print(f"The temperature in Fahrenheit is: {fahrenheit}")
```

#####  Calcule o salário bruto (horas * salario por hora)

```python
value_hour = float(input("Please enter how much you earn per hour: "))
hour = float(input("Please enter how many hours you work per day: "))
gross_salary = (hour * value_hour)

print(f'Gross salary per hour is: {gross_salary}')
```

##### Calcule o desconto do IR (11% do salário bruto)

```python
value_hour = float(input("Please enter how much you earn per hour: "))
hour = float(input("Please enter how many hours you work per day: "))
gross_salary = (hour * value_hour)
income_tax = (gross_salary * 0.11)

print(f'Your income tax is: {income_tax}')
```

##### Calcule o desconto do INSS (8% do salário bruto)

```python
value_hour = float(input("Please enter how much you earn per hour: "))
hour = float(input("Please enter how many hours you work per day: "))
gross_salary = (hour * value_hour)
inss_tax = (gross_salary * 0.08)

print(f'Your INSS tax is: {inss_tax}')
```

##### Calcule o desconto do sindicato (5% do salário bruto)

```python
value_hour = float(input("Please enter how much you earn per hour: "))
hour = float(input("Please enter how many hours you work per day: "))
gross_salary = (hour * value_hour)

sindicate_tax = (gross_salary * 0.05)

print(f'Your sindicate tax is: {sindicate_tax}')
```

##### Calcule o salário líquido (salário bruto - descontos)

```python
value_hour = float(input("Please enter how much you earn per hour: "))
hour = float(input("Please enter how many hours you work per day: "))
gross_salary = (hour * value_hour)
sindicate_tax = (gross_salary * 0.05)
inss_tax = (gross_salary * 0.08)
income_tax = (gross_salary * 0.11)

liquid_salary = (gross_salary - sindicate_tax - inss_tax - income_tax)

print(f'Your liquid salary is: {liquid_salary}')
```
