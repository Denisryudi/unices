#### Consulta de informações de um CEP

```python
import requests

cep= input("Digite o CEP: ")

cep = cep.replace("-", "").replace(".", "")
if len(cep) == 8:
    link = f'https://viacep.com.br/ws/{cep}/json/'
    
    requisicao = requests.get(link)
    print(requisicao)
    
    dic_requisicao = requisicao.json()
    
    logradouro = dic_requisicao['logradouro']
    cidade = dic_requisicao['localidade']
    uf = dic_requisicao['uf']
    
    print(uf)
else:
    print('CEP inválido, favor verificar se tem 8 dígitos')
```

```python

uf = 'RJ'
cidade = 'Rio de Janeiro'
logradouro = 'Rio Branco'
link1 = f'https://viacep.com.br/ws/{uf}/{cidade}/{logradouro}/json/'
requisicao = requests.get(link1)
print(requisicao)
dic_requisicao = requisicao.json()
print(dic_requisicao[0]['cep'])
```
