```python
#proto

import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
marcadores = np.where(vendas == np.max(vendas), 'x', '')

#paleta cpres
cores = plt.get_cmap('Pastel2').colors
cycler_colors = cycler('color', cores)
plt.rc('axes', prop_cycle=cycler_colors)

#gráfico base
plt.plot(meses, vendas, marker='o', linestyle='--', linewidth=2)

#marcadores e anotações
for posicao, marcador in enumerate(marcadores):
    if marcador == 'x':
        cor='green'
    else:
        cor='blue'
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, color=cor)

# Configurações de estilo de texto
texto_estilo = {'fontsize': 10, 'fontweight': 'bold'}
    
plt.annotate('Peak', color='red', xy=(11, 108), xytext=(11, 103), arrowprops=dict(facecolor='purple', width=1.8, headwidth=8, shrink=0.05), **texto_estilo)

#destaque de região ax= eixo v= vertical span = em uma região . alpha= transparência  
plt.axvspan(6, 11, alpha=0.2, color=cores[2])

plt.text(6, 80, 'Recuperação de crescimento')

plt.box(False)

plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano', color=cores[4], fontsize=15, weight='bold')
plt.grid(True)
plt.show()

print(marcadores)
```

```python

import matplotlib.pyplot as plt
import numpy as np

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

# usando o np where, crie uma lista de marcadores x para o valor mínimo e o para o resto
marcadores = np.where(vendas == np.min(vendas), 'x', '')

plt.plot(meses, vendas)

for posicao, marcador in enumerate(marcadores):
    if marcador == 'x':
        cor = 'red'
        tamanho = 80
    else:
        cor = 'blue'
        tamanho = 0
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, s=tamanho, color=cor)

# adicionando uma anotação para mostrar o menor valor em março
plt.annotate('Menor Valor', color='red', xy=(2.15, 78), xytext=(3, 80), arrowprops=dict(facecolor='red'))



plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
marcadores = np.where(vendas == np.max(vendas), 'x', '')

#paleta cpres
cores = plt.get_cmap('Pastel2').colors
cycler_colors = cycler('color', cores)
plt.rc('axes', prop_cycle=cycler_colors)

#gráfico base
plt.plot(meses, vendas, marker='o', linewidth=2)

# Configurações de estilo de texto
texto_estilo = {'fontsize': 9}

#marcadores e anotações
for posicao, marcador in enumerate(marcadores):
    if marcador == 'x':
        cor='green'
    else:
        cor='teal'
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, color=cor)
    plt.annotate(vendas[posicao], color=cor, xy=(meses[posicao], vendas[posicao]), xytext=(-5, 5), textcoords='offset points', **texto_estilo)
    



#destaque de região ax= eixo v= vertical span = em uma região . alpha= transparência  
plt.axvspan(6, 11, alpha=0.2, color=cores[2])

plt.text(6, 80, 'Recuperação de crescimento')

plt.box(False)
plt.tick_params(axis='both', which='major')
plt.yticks([])

plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano', color=cores[4], fontsize=15, weight='bold')
plt.grid(True)
plt.show()

print(marcadores)
```
