## Gráfico de barras empilhadas para mostrar a distribuição de diferentes tipos de vendas ao longo do tempo

Considere que você tenha uma empresa e queira mostrar a distribuição de diferentes tipos de vendas ao longo do tempo. Crie um gráfico de barras empilhadas:

```python
import matplotlib.pyplot as plt
from cycler import cycler

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

cores_cycler = cycler('color', plt.get_cmap('Pastel2').colors)
plt.rc('axes', prop_cycle=cores_cycler)

plt.bar(anos, vendas_online, label='Online')
plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

plt.xticks(ticks=anos, labels=['Ano 1', 'Ano 2', 'Ano 3', 'Ano4'])
plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões de reais')
plt.title('Vendas Online vs Lojas')

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

# rc = run configuration (configuração quando executa)
plt.rc('axes', prop_cycle=(cycler('color', plt.get_cmap('Pastel2').colors)))
 
plt.bar(anos, vendas_online, label='Online')
plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

# forçar os anos a serem exibidos como inteiros
plt.xticks(ticks=anos)

plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show()
```
